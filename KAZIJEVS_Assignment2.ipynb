In this assingment, you will analyze local and global temperature data and compare the temperature trends of some cities to global temperature trends using pandas dataframes. Given the dataset for the annual average temperatures for the cities, find the city, the average temperature distribution of which was the most similar to the global average temperature distribution.
```python
#include all the modules here
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
import math

import warnings
warnings.filterwarnings('ignore')
```

```python
# load in the dataset into a pandas dataframe and show the size of the file and show the first 5 rows of the data
temp_data = pd.read_csv(r'temp.csv')
print(temp_data.shape)
temp_data.head()
```

```python
# Create a new data frame with the data of these cities only: 'Global', 'Nashville', 'Milan', 'Paris', 'New York', 'Moscow', 'Stockholm'
subset_temp_data = temp_data[temp_data["city"].isin(['Global', 'Nashville', 'Milan', 'Paris', 'New York', 'Moscow', 'Stockholm'])]
```

```python
# Take only the data for the years between 1759 and 2013
subset_temp_data = subset_temp_data[(subset_temp_data["year"] >= 1759) & (subset_temp_data["year"] <= 2013)]
print(subset_temp_data)
```

```python
# Format your frame such that index will be the years, columns will be the cities and values will be the avg_temp
temp_data_wide = subset_temp_data.pivot_table(index='year', columns='city', values='avg_temp')
temp_data_wide.head()
```

```python
# see the first 40 elements of the data, see some NaN values
temp_data_wide.head(40)
```

```python
# Fill the null data with the previous value
temp_data_wide= temp_data_wide.fillna(method='pad')
```

```python
# see the first 40 elements of the data, there shouldn't be nan valules
temp_data_wide.head(40)
```

```python
# Draw all the global and city temperatures on the same plot
plt.rcParams["figure.figsize"] = [14,8]

plt.plot(temp_data_wide.index, temp_data_wide["Global"],'r', label="Global")
plt.plot(temp_data_wide.index, temp_data_wide["Nashville"],'b', label="Nashville")
plt.plot(temp_data_wide.index, temp_data_wide["Milan"],'g', label="Milan")
plt.plot(temp_data_wide.index, temp_data_wide["Paris"],'c', label="Paris")
plt.plot(temp_data_wide.index, temp_data_wide["New York"],'m', label="New York")
plt.plot(temp_data_wide.index, temp_data_wide["Moscow"],'k', label="Moscow")
plt.plot(temp_data_wide.index, temp_data_wide["Stockholm"],'y', label="Stockholm")


plt.xlabel('years', fontsize=14)
plt.ylabel('Average Temperature', fontsize=14)
plt.title('City Temperatures for years between 1759 and 2013', fontsize=14)
plt.rcParams["legend.shadow"]
plt.legend(loc='best')
```

```python
# Find the correlation matrix
temp_data_wide.corr()
```

```python
# Observe the correlation between cities and global temp
g = sb.regplot(data=temp_data_wide, x= temp_data_wide["Global"], y=temp_data_wide["Nashville"])
g.figure.set_size_inches(5, 5)
plt.show()

g = sb.regplot(data=temp_data_wide, x= temp_data_wide["Global"], y=temp_data_wide["Moscow"])
g.figure.set_size_inches(5, 5)
plt.show()
```

### (60 Points) Second part the assignment, use rolling window caluculations and calculate the correlations again.

```python
# <<your data frame>> = <<your data frame>>.rolling(window=7,min_periods=1).mean()
rolled_data = temp_data_wide.rolling(window=7,min_periods=1).mean()
```

```python
# Draw all the global and city temperatures on the same plot using the rolled data for Global,Milan,Moscow,Nashville,New York,Paris,Stockholm
plt.figure(figsize=(18,8))

plt.plot(rolled_data.index, rolled_data["Global"],'r', label="Global")
plt.plot(rolled_data.index, rolled_data["Nashville"],'b', label="Nashville")
plt.plot(rolled_data.index, rolled_data["Milan"],'g', label="Milan")
plt.plot(rolled_data.index, rolled_data["Paris"],'c', label="Paris")
plt.plot(rolled_data.index, rolled_data["New York"],'m', label="New York")
plt.plot(rolled_data.index, rolled_data["Moscow"],'k', label="Moscow")
plt.plot(rolled_data.index, rolled_data["Stockholm"],'y', label="Stockholm")

plt.xlabel("years")
plt.ylabel("Average Temperature")
plt.title("City temperatures for years between 1759 and 2013")
plt.plot(rolled_data)
plt.legend(list(rolled_data.columns),loc='best')
```

```python
# Find the correlation matrix for the rolled data and tell which one has the highest correlation with global temp
corr = rolled_data.corr()
largest_corr = corr.loc['Global'].nlargest(2)
print("Highest correlation with Global is",largest_corr.index[1],":",largest_corr[1])
corr
```

```python
# Original df
temp_data.head()
```

```python
# Pivot table
# We will take all the years this time because some cities have data
pivot_df = temp_data.pivot_table(index='year', columns='city', values='avg_temp')
pivot_df.head()
```

```python
# Fill the null data with the previous value
# bfill - backward fill (some countries doesn't have intitial value)
pivot_df = pivot_df.fillna(method='bfill')
pivot_df.head()
```

```python
#Rolling window to get average
df_roll = pivot_df.rolling(window=7,min_periods=1).mean()
#Getting correlation with Global
global_corr = df_roll.corr()['Global']
#Finding 2 largest values (1st - Global with itself; 2nd - Global with other country)
largest = global_corr.nlargest(n=2, keep='first')
#Printing
print("Highest correlation: ", largest.index[1], "\nValue: ", largest[1])
```
